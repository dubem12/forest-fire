```python
def get_season(month):
    # Define a mapping of months to seasons
    season_mapping = {
        "jan": "winter",
        "feb": "winter",
        "mar": "spring",
        "apr": "spring",
        "may": "spring",
        "jun": "summer",
        "jul": "summer",
        "aug": "summer",
        "sep": "autumn",
        "oct": "autumn",
        "nov": "autumn",
        "dec": "winter",
    }

    # Convert the input to lowercase to handle case variations
    month_lower = month.lower()

    # Check if the month is in the mapping, if not, return None
    return season_mapping.get(month_lower, None)


# Apply the get_season function to create a new column 'season'
df["season"] = df["month"].apply(get_season)

# Display the DataFrame with the new 'season' column
print(df)
```

for days of the week we will factor in weekend and weekdays, as it is more likely that an unattended forest area will be more likely to burn longer than an actively watched forest, of course we will have a problem as we won´t consider office hours . Friday 9pm is almost same as saturday 12pm



```python
def categorize_day(day):
    if day in ["sat", "sun"]:
        return "weekend"
    else:
        return "weekday"


# Apply the categorize_day function to create a new column 'day_category'
df["day_category"] = df["day"].apply(categorize_day)
```
